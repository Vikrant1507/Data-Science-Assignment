## Q1) Delivery_time -> Predict delivery time using sorting time. Build a simple linear regression model by performing EDA and do necessary transformations and select the best model using R or Python.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv('D:\\delivery_time.csv')
dataset
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
sns.distplot(dataset['Delivery Time'])
plt.show()
```

```python
sns.distplot(dataset['Sorting Time'])
plt.show()
```

### Feature Engineering

```python
# Renaming Columns
dataset=dataset.rename({'Delivery Time':'delivery_time', 'Sorting Time':'sorting_time'},axis=1)
dataset
```

```python
#corelation analysis
dataset.corr()
```

```python
sns.regplot(x=dataset['sorting_time'],y=dataset['delivery_time'])
plt.show()
```

```python
#model Building
model=smf.ols("delivery_time~sorting_time",data=dataset).fit()
```

### model testing

```python
model.params
```

```python
# Finding tvalues and pvalues
model.tvalues , model.pvalues
```

```python
# Finding Rsquared Values
model.rsquared , model.rsquared_adj
```

### model prediction

```python
# Manual prediction for say sorting time 5
delivery_time = (6.582734) + (1.649020)*(5)
delivery_time
```

```python
# Automatic Prediction for say sorting time 5, 8
new_data=pd.Series([5,8])
new_data
```

```python
data_pred=pd.DataFrame(new_data,columns=['sorting_time'])
data_pred
```

```python
model.predict(data_pred)
```

## Q2) Salary_hike -> Build a prediction model for Salary_hike Build a simple linear regression model by performing EDA and do necessary transformations and select the best model using R or Python.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
dataset1=pd.read_csv('D:\\Salary_Data.csv')
dataset1
```

```python
dataset1.describe()
```

### EDA AND DATA VISUALIZATION

```python
dataset1.info()
```

```python
sns.distplot(dataset1['YearsExperience'])
plt.show()
```

```python
sns.distplot(dataset1['Salary'])
plt.show()
```

### Corelation Analysis

```python
dataset1.corr()
```

```python
sns.regplot(x=dataset1['YearsExperience'],y=dataset1['Salary'])
plt.show()
```

### Model Building

```python
model=smf.ols("Salary~YearsExperience",data=dataset1).fit()
```

### model Testing

```python
# Finding Cefficient Parameters
model.params
```

```python
model.tvalues,model.pvalues
```

```python
model.rsquared , model.rsquared_adj
```

### Model Predictions

```python
# Manual prediction for say 3 Years Experience
Salary = (25792.200199) + (9449.962321)*(3)
Salary
```

```python
# Automatic Prediction for say 3 & 5 Years Experience 
```

```python
new_data=pd.Series([3,5])
new_data
```

```python
data_pred=pd.DataFrame(new_data,columns=['YearsExperience'])
data_pred
```

```python
model.predict(data_pred)
```
